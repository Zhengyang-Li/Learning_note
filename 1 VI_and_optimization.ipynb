## An equilibrium problem
There are two commodities indexed by $i$, e.g., 1: sprite and 2: coke. Assume that the demand for sprite and coke are two decreasing functions $d_1(p)$ and $d_2(p)$ with respect to price $p$; the supply for the sprite and coke are two increasing functions $s_1(p)$ and $s_2(p)$ with respect to price $p$, where
$$
p=\left(\begin{array}{l}
p_{1} \\
p_{2}
\end{array}\right)
$$
is the price vector.

We group the aggregate demands for the commodities into the 2-dimensional column vector
$$
d(p)=\left(\begin{array}{l}
d_{1}(p) \\
d_{2}(p)
\end{array}\right)
$$

and the aggregate supplies of the commodities into the 2-dimensional column vector 
$$
s(p)=\left(\begin{array}{l}
s_{1}(p) \\
s_{2}(p)
\end{array}\right)
$$

What is the demand, supply, and price under different conditions?

### Market equilibrium with equalities only (systems of equations)
The market equilibrium conditions that require that the supply of each commodity must be equal to the demand for each commodity at the equilibrium price vector $p^*$:
$$
s\left(p^{*}\right)-d\left(p^{*}\right)=0
$$

#### VI formulation
Let $F(p)=s(p)-d(p)$ and $x=p$, this problem can be formulated as a $VI(F,\mathbb{R}^n)$, i.e.,
$$
F\left(x^{*}\right)^{T} \cdot\left(x-x^{*}\right) \geq 0, \quad \forall x \in \mathbb{R}^n
$$

#### Proof
If $F\left(x^{*}\right)=0$, then the condition holds. Conversely, if $x^*$ satisfy the above equation, let $x=x^*-F(x^*)$, we have
$$
F\left(x^{*}\right)^{T} \cdot\left(-F(x^{*})\right) \geq 0 \Rightarrow -||F\left(x^{*}\right)||^2 \geq 0
$$
Therefore, $F\left(x^{*}\right)=0$.

Note that $x \in \mathbb{R}^n$ is allowed to take negative value. Therefore, the system of equations is not suitable in the case of nonnegativity assumptions on certain variables such as prices.

### Market equilibrium conditions (system of equalities and inequalities)
Assume that the prices must now be nonnegative in the market equilibrium example presented earlier.
Hence, we consider the following situation, in which the demand function are given as previously as are the supply functions, but now, instead of the market equilibrium conditions, which are represented by a system of equations, we have the following equilibrium conditions.
$$
s_{i}\left(p^{*}\right)-d_{i}\left(p^{*}\right) 
\begin{cases}
=0, & \text { if } p_{i}^{*}>0 \\ 
\geq 0, & \text { if } p_{i}^{*}=0
\end{cases} \quad \forall i=1,2
$$
- These equilibirum condition state that if the price of a commodity is positive in equilibrium then the supply of that commodity must be equal to the demand for that commodity. 
- If the price of a commodity at equilibrium is zero, then there may be excess supply of that commodity at equilibrium, that is, $s_{i}\left(p^{*}\right)-d_{i}\left(p^{*}\right)>0$, or the market clears.
- This system of equalities and inequalities guarantees that the prices of the instruments do not take on negative values, which may occur in the system of equations expressing the market clearing conditions.

#### VI formulation
Let $F(p)=s(p)-d(p)$ and $x=p$, this problem can be formulated as a $VI(F,\mathbb{R}_+^n)$, i.e.,
$$
F\left(x^{*}\right)^{T} \cdot\left(x-x^{*}\right) \geq 0, \quad \forall x \in \mathbb{R}_+^n
$$


## A special case: demand functions and supply functions that are separable
That is, the Jacobians of $s(p)$ and $d(p)$ are diagonal:
$$
\nabla s(p)=\left(\begin{array}{ccccc}
\frac{\partial s_{1}}{\partial p_{1}} & 0 & 0 & \ldots & 0 \\
0 & \frac{\partial s_{2}}{\partial p_{2}} & 0 & \ldots & 0 \\
\vdots & \ddots & \ddots & \ddots & \vdots \\
0 & 0 & 0 & \ldots & \frac{\partial s_{n}}{\partial p_{n}}
\end{array}\right)
$$
$$
\nabla d(p)=\left(\begin{array}{ccccc}
\frac{\partial d_{1}}{\partial p_{1}} & 0 & 0 & \ldots & 0 \\
0 & \frac{\partial d_{2}}{\partial p_{2}} & 0 & \ldots & 0 \\
\vdots & \ddots & \ddots & \ddots & \vdots \\
0 & 0 & 0 & \ldots & \frac{\partial d_{n}}{\partial p_{n}}
\end{array}\right)
$$

In this special case, the price vector $p^*$ that satisfies the equilibrium conditions can be obtained by solving the following optimization problem:
$$
\min \sum_{i=1} \int_{0}^{p_{i}} s_{i}(x) d x-\sum_{i=1} \int_{0}^{p_{i}} d_{i}(y) d y
$$
$$
\text{s.t. } p_i \geq 0 \quad \forall i
$$

### An example
Demand function:
$$
\nabla d(p)=\left(\begin{array}{cc}
-2 & 0 \\
0 & -2
\end{array}\right)
\Rightarrow
d(p)=\left(\begin{array}{l}
-2p_1 + 10 \\
-2p_2 + 10
\end{array}\right)
\Rightarrow
\int_0^p d(v)dv=\left(\begin{array}{cc}
-p_1^2+10p_1 \\
-p_2^2+10p_2
\end{array}\right)
$$

Supply function:
$$
\nabla s(p)=\left(\begin{array}{cc}
2 & 0 \\
0 & 2
\end{array}\right)
\Rightarrow
s(p)=\left(\begin{array}{l}
2p_1 \\
2p_2
\end{array}\right)
\Rightarrow
\int_0^p s(v)dv=\left(\begin{array}{cc}
p_1^2 \\
p_2^2
\end{array}\right)
$$

Therefore, the optimization model:
$$
\min p_1^2 + p_2^2 -(-p_1^2+10p_1)-(-p_2^2+10p_2)
$$
$$
\text{s.t. } p_i \geq 0 \quad \forall i=1,2
$$

Use CVXPY to solve this optimization model.

```python
import cvxpy as cp

p = cp.Variable(2, nonneg=True)

# Objective function
obj = cp.Minimize(p[0]**2 + p[1]**2 - (-p[0]**2 + 10*p[0]) - (-p[1]**2 + 10*p[1]))

# Constraints
constraints = []

prob = cp.Problem(obj, constraints)
prob.solve()

#output results
print('Obj. Val.', prob.value)
print('Price', p.value)
print('Spply', 2*p.value)
print('Demand', -2*p.value+10)
```

```python
# The objective function surface
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.01)
y = x
xaxis, yaxis = np.meshgrid(x, y)
z = xaxis**2 + yaxis**2 - (-xaxis**2 + 10*xaxis) - (-yaxis**2 + 10*yaxis)

fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(xaxis, yaxis, z)
plt.xlabel('p1')
plt.ylabel('p2')
plt.show()
```

```python
# Graphic interpretation of equilibrium

plt.rcParams['figure.figsize'] = [15, 5]

# equilibrium for commodity 1 (sprite)
p1 = np.arange(0, 5, 0.01)
d1 = -2 * p1 + 10
s1 = 2 * p1

ax1 = plt.subplot(1,2,1)
ax1.plot(p1, d1, label='demand')
ax1.plot(p1, s1, label='supply', color='k')

p1 = np.arange(0, p.value[0], 0.01)
d1 = -2 * p1 + 10
s1 = 2 * p1
ax1.fill_between(p1, d1, s1, color='r', alpha=0.5, label='area=12.25')
ax1.legend()

plt.xlabel('demand')
plt.ylabel('price')
plt.title("Commodity 1")

# equilibrium for commodity 2 (coke)
p2 = np.arange(0, 5, 0.01)
d2 = -2 * p2 + 10
s2 = 2 * p2

ax2 = plt.subplot(1,2,2)
ax2.plot(p2, d2, label='demand')
ax2.plot(p2, s2, label='supply', color='k')

p2 = np.arange(0, p.value[1], 0.01)
d2 = -2 * p2 + 10
s2 = 2 * p2
ax2.fill_between(p2, d2, s2, color='r', alpha=0.5, label='area=12.25')
ax2.legend()

plt.xlabel('demand')
plt.ylabel('price')
plt.title("Commodity 2")
plt.show()
```

# The relation between optimization problems (MP) and variational inequalities (VI)

## From MP to VI
MP:
$$
\min_{x \in K} f(x)
$$
Solution $x*$.

If 
$$
f \leftarrow \text{continuously differentiable}
$$
$$
K \leftarrow \text{closed and convex}
$$

Then, $x^$ is a solution of the variational inequality problem:
$$
\nabla f\left(x^{*}\right)^{T} \cdot\left(x-x^{*}\right) \geq 0, \quad \forall x \in K
$$

## From VI to MP
VI:
$$
F\left(x^{*}\right)^{T} \cdot\left(x-x^{*}\right) \geq 0, \quad \forall x \in K
$$
If $\nabla F$ is symmetric and positive definite, then this VI can be write as a MP.
$$
\min_{x \in K} f(x)
$$
where $\nabla f(x)=F(x)$.

### Why?
- If $\nabla F$ is sysmetric and positive definite:
$$
\nabla F = 
\left(
\begin{array}{l}
2 & 1 \\
1 & 2
\end{array}\right)
\Rightarrow
F =
\left(
\begin{array}{l}
2x_1 + x_2 \\
x_1 + 2x_2
\end{array}\right)
\Rightarrow
f(x) = x_1^2 + x_1 x_2 + x_2^2
$$
$$
\nabla F\text{ positive definite}
\Rightarrow
F \text{ increasing function}
\Rightarrow
f(x) \text{ convex function}
$$

- If $\nabla F$ is not sysmetric:
$$
\nabla F = 
\left(
\begin{array}{l}
2 & 1 \\
0 & 2
\end{array}\right)
\Rightarrow
F =
\left(
\begin{array}{l}
2x_1 + x_2 \\
0x_1 + 2x_2
\end{array}\right)
\Rightarrow
f(x) \text{ does not exist!}
$$

- If $\nabla F$ is symmetric but negative definite:
$$
\nabla F = 
\left(
\begin{array}{l}
1 & 2 \\
2 & 1
\end{array}\right)
\Rightarrow
F =
\left(
\begin{array}{l}
x_1 + 2x_2 \\
2x_1 + x_2
\end{array}\right)
\Rightarrow
f(x) = \frac{1}{2}x_1^2 + 2x_1 x_2 + \frac{1}{2}x_2^2 (\text{ Not convex, a saddle surface!})
$$

#### What about $\nabla F$ is semmetric and semi-positive definite?
Multiple solutions.

```python
# Condition 1: a convex function
plt.rcParams['figure.figsize'] = [9, 6]

x = np.arange(-2, 2, 0.01)
y = np.arange(-2, 2, 0.01)
xaxis, yaxis = np.meshgrid(x, y)
z = xaxis ** 2 + xaxis * yaxis + yaxis ** 2

fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(xaxis, yaxis, z)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
# Condition 3: a saddle surface
x = np.arange(-2, 2, 0.01)
y = np.arange(-10, 10, 0.01)
xaxis, yaxis = np.meshgrid(x, y)
z = 0.5 * xaxis ** 2 + 2 * xaxis * yaxis + 0.5 * yaxis ** 2

fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(xaxis, yaxis, z)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

### Consider a special case
- If $\nabla F$ is sysmetric but not neccessarily positive definite (variable dependent):
$$
\nabla F = 
\left(
\begin{array}{l}
2 & 2x_2 \\
2x_2 & 2
\end{array}\right)
\Rightarrow
F =
\left(
\begin{array}{l}
2x_1 + x_2^2 \\
2x_1x_2 + 2x_2
\end{array}\right)
\Rightarrow
f(x) = x_1^2 + x_1 x_2^2 + x_2^2
$$
When $x_2\in [-1,1]$, $\nabla F$ is positive definite. Thus, $f(x)$ is convex in $x_2\in [-1,1]$.

```python
# The whole solution space
x = np.arange(-5, 5, 0.01)
y = x
xaxis, yaxis = np.meshgrid(x, y)
z = xaxis ** 2 + xaxis * yaxis ** 2 + yaxis ** 2

fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(xaxis, yaxis, z)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
# The space where 1 <= x_2 <= 1
x = np.arange(-1, 1, 0.01)
y = x
xaxis, yaxis = np.meshgrid(x, y)
z = xaxis ** 2 + xaxis * yaxis ** 2 + yaxis ** 2

fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(xaxis, yaxis, z)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
